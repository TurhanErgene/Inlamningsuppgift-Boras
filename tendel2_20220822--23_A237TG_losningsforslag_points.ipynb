```python
#1p
def minstaTal(lista):
    minVarde = None #kan sättas till ett "jättestort" tal som start, men detta är bättre
    for tal in lista:
        if (minVarde is None or tal < minVarde):
            minVarde = tal
    return minVarde

#1p
def storstaTal(lista):
    maxVarde = None
    for tal in lista:
        if (maxVarde is None or tal > maxVarde):
            maxVarde = tal
    return maxVarde

#1p
def medelVarde(lista):
    sumAllaTal = 0
    for i in range(len(lista)):
        sumAllaTal += lista[i]
    return sumAllaTal/len(lista)
    #alt. sum(lista)/len(lista)

#1p
import random
slumplista = []
for i in range(0,5):
    slumplista.append(random.randint(1,10))
#alt. slumplista = random.sample(range(1, 10), 5)
print("Slumplistan blev: ")
print(slumplista)
print()

print("Minsta talet: " + str(minstaTal(slumplista)))
print("Största talet: "+ str(storstaTal(slumplista)))
print("Medelvärde: " + str(medelVarde(slumplista)))
```

```python
ANTAL_KOLUMNER = 5
tal = 2
raknare = 0

antal_primtal = int(input("Skriv in antal primtal som önskas visas: "))
print(f'De första {antal_primtal} primtalen är: ')

while raknare < antal_primtal:
    ar_prim = True #False --> True
    divisor = 2
    while divisor <= tal / 2: #>= --> <=
        if tal % divisor == 0: #// --> %
            ar_prim = False 
            break #continue --> break
        divisor += 1
    if ar_prim:
        raknare += 1 #-= --> +=
        print(f' {tal:5d}', end = '')
        if raknare % ANTAL_KOLUMNER == 0:
            print()
    tal += 1
```
